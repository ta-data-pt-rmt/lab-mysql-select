```python
# Create a database
create_db_statement = "CREATE DATABASE IF NOT EXISTS Publications2;"
cursor = cnx.cursor()
cursor.execute(create_db_statement)

# Switch to the created database
use_db_statement = "USE Publications2;"
cursor.execute(use_db_statement)

# Dataframe: authors
create_table_authors = """
CREATE TABLE IF NOT EXISTS Authors (
    au_id VARCHAR(15) PRIMARY KEY,
    au_lname VARCHAR(255),
    au_fname VARCHAR(255),
    phone VARCHAR(20),
    address VARCHAR(255),
    city VARCHAR(255),
    state VARCHAR(2),
    zip VARCHAR(10),
    contract INT
);
"""
cursor.execute(create_table_authors)
print("Table 'Authors' created successfully.")

# Dataframe: discounts
create_table_discounts = """
CREATE TABLE IF NOT EXISTS Discounts (
    discounttype VARCHAR(255),
    stor_id INT,
    lowqty INT,
    highqty INT,
    discount FLOAT
);
"""
cursor.execute(create_table_discounts)
print("Table 'Discounts' created successfully.")

# Dataframe: employee
create_table_employee = """
CREATE TABLE IF NOT EXISTS Employee (
    emp_id VARCHAR(255) PRIMARY KEY,
    fname VARCHAR(255),
    minit VARCHAR(1),
    lname VARCHAR(255),
    job_id INT,
    job_lvl INT,
    pub_id INT,
    hire_date DATE
);
"""
cursor.execute(create_table_employee)
print("Table 'Employee' created successfully.")

# Dataframe: jobs
create_table_jobs = """
CREATE TABLE IF NOT EXISTS Jobs (
    job_id INT PRIMARY KEY,
    job_desc VARCHAR(255),
    min_lvl INT,
    max_lvl INT
);
"""
cursor.execute(create_table_jobs)
print("Table 'Jobs' created successfully.")

# Dataframe: pub_info
create_table_pub_info = """
CREATE TABLE IF NOT EXISTS Pub_Info (
    pub_id INT PRIMARY KEY,
    logo VARCHAR(2000),
    pr_info TEXT
);
"""
cursor.execute(create_table_pub_info)
print("Table 'Pub_Info' created successfully.")

# Dataframe: publishers
create_table_publishers = """
CREATE TABLE IF NOT EXISTS Publishers (
    pub_id INT PRIMARY KEY,
    pub_name VARCHAR(255),
    city VARCHAR(255),
    state VARCHAR(255),
    country VARCHAR(255)
);
"""
cursor.execute(create_table_publishers)
print("Table 'Publishers' created successfully.")

# Dataframe: roysched
create_table_roysched = """
CREATE TABLE IF NOT EXISTS Roysched (
    title_id VARCHAR(255),
    lorange INT,
    hirange INT,
    royalty FLOAT
);
"""
cursor.execute(create_table_roysched)
print("Table 'Roysched' created successfully.")

# Dataframe: sales
create_table_sales = """
CREATE TABLE IF NOT EXISTS Sales (
    stor_id INT,
    ord_num VARCHAR(255),
    ord_date DATE,
    qty INT,
    payterms VARCHAR(255),
    title_id VARCHAR(255)
);
"""
cursor.execute(create_table_sales)
print("Table 'Sales' created successfully.")

# Dataframe: stores
create_table_stores = """
CREATE TABLE IF NOT EXISTS Stores (
    stor_id INT PRIMARY KEY,
    stor_name VARCHAR(255),
    stor_address VARCHAR(255),
    city VARCHAR(255),
    state VARCHAR(255),
    zip VARCHAR(10)
);
"""
cursor.execute(create_table_stores)
print("Table 'Stores' created successfully.")

# Dataframe: titleauthor
create_table_titleauthor = """
CREATE TABLE IF NOT EXISTS TitleAuthor (
    au_id VARCHAR(255),
    title_id VARCHAR(255),
    au_ord INT,
    royaltyper INT
);
"""
cursor.execute(create_table_titleauthor)
print("Table 'TitleAuthor' created successfully.")

# Dataframe: titles
create_table_titles = """
CREATE TABLE IF NOT EXISTS Titles (
    title_id VARCHAR(255) PRIMARY KEY,
    title VARCHAR(255),
    type VARCHAR(255),
    pub_id INT,
    price FLOAT,
    advance FLOAT,
    royalty FLOAT,
    ytd_sales FLOAT,
    notes TEXT,
    pubdate DATE
);
"""
cursor.execute(create_table_titles)
print("Table 'Titles' created successfully.")
```

```python
# Connect to MySQL server
cnx = mysql.connector.connect(user='root', password='********', port=3306, database='Publications2')
cursor = cnx.cursor()

# Define a dictionary with table names as keys and dataframes as values
dataframes = {
    'Authors': authors,
    'Discounts': discounts,
    'Employee': employee,
    'Jobs': jobs,
    'Pub_info': pub_info,
    'Publishers': publishers,
    'Roysched': roysched,
    'Sales': sales,
    'Stores': stores,
    'TitleAuthor': titleauthor,
    'Titles': titles
}

# Iterate through the dictionary and insert data into MySQL
for table_name, df in dataframes.items():
    # Create the table
    create_table_statement = f"CREATE TABLE IF NOT EXISTS {table_name} ({', '.join([f'{col} VARCHAR(255)' for col in df.columns])});"
    cursor.execute(create_table_statement)

    # Insert data into the table
    for _, row in df.iterrows():
        placeholders = ', '.join(['%s' for _ in df.columns])
        insert_data_statement = f"INSERT INTO {table_name} VALUES ({placeholders})"
        cursor.execute(insert_data_statement, tuple(row))

    print(f"{len(df)} rows inserted into '{table_name}' table.")

# Commit changes and close connection
cnx.commit()
cursor.close()
cnx.close()
```
